```python
#Bài 13 Viết hàm trộn 2 mảng một chiều thành 1 mảng một chiều
#với mỗi phần tử của mảng mới là tổng của 2 phần tương ứng từ 2 mảng cho trước
#Trong quá trình trộn 2 mảng nếu mảng nào còn phần tử thì các phần tử còn lại của mảng đó sẽ đưa vào mảng mới

def tron_hai_mang_va_tinh_tong(mang1, mang2):
    mang_ket_qua = []
    do_dai_mang1 = len(mang1)
    do_dai_mang2 = len(mang2)
    do_dai_lon_nhat = max(do_dai_mang1, do_dai_mang2)


    for i in range(do_dai_lon_nhat):
        if i < do_dai_mang1 and i < do_dai_mang2:
            mang_ket_qua.append(mang1[i] + mang2[i])
        elif i < do_dai_mang1:
            mang_ket_qua.append(mang1[i])
        elif i < do_dai_mang2:
            mang_ket_qua.append(mang2[i])
    return mang_ket_qua
# Các ví dụ sử dụng hàm với dữ liệu cố định:
arr_a_ex = [1, 2, 3, 4, 5]
arr_b_ex = [10, 20, 30]
result1 = tron_hai_mang_va_tinh_tong(arr_a_ex, arr_b_ex)
print(f"\nVí dụ 1:")
print(f"  Mảng A: {arr_a_ex}")
print(f"  Mảng B: {arr_b_ex}")
print(f"  Mảng kết quả (trộn A và B): {result1}")
```

```python
#Bài 14 Cho mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:
import numpy as np
# a) Tạo mảng a chứa các số nguyên ngẫu nhiên.
def tao_mang_ngau_nhien_2_chieu(m, n, low=0, high=100):
    """
    Tạo mảng 2 chiều chứa các số nguyên ngẫu nhiên.
    m: số dòng
    n: số cột
    low: giá trị nhỏ nhất của số ngẫu nhiên (bao gồm)
    high: giá trị lớn nhất của số ngẫu nhiên (không bao gồm)
    """
    return np.random.randint(low, high, size=(m, n))

# Ví dụ sử dụng:
m = 3 # số dòng
n = 4 # số cột
array_a = tao_mang_ngau_nhien_2_chieu(m, n)
print(f"--- a) Mảng 2 chiều a với {m} dòng, {n} cột chứa số nguyên ngẫu nhiên: ---\n{array_a}\n")

# b) Xuất các phần tử thuộc dòng k.
def xuat_dong_k(mang_2d, k):
    """
    Xuất các phần tử thuộc dòng k của mảng 2 chiều.
    mang_2d: mảng 2 chiều.
    k: chỉ số dòng cần xuất (bắt đầu từ 0).
    """
    if 0 <= k < mang_2d.shape[0]:
        print(f"Các phần tử thuộc dòng {k} là: {mang_2d[k]}")
    else:
        print(f"Chỉ số dòng {k} không hợp lệ. Mảng có {mang_2d.shape[0]} dòng.")

print("--- b) Xuất các phần tử thuộc dòng k: ---")
k_dong = 1 # Dòng thứ 2 (chỉ số 1)
xuat_dong_k(array_a, k_dong)
print()

# c) Xuất các phần tử thuộc cột k.
def xuat_cot_k(mang_2d, k):
    """
    Xuất các phần tử thuộc cột k của mảng 2 chiều.
    mang_2d: mảng 2 chiều.
    k: chỉ số cột cần xuất (bắt đầu từ 0).
    """
    if 0 <= k < mang_2d.shape[1]:
        print(f"Các phần tử thuộc cột {k} là: {mang_2d[:, k]}")
    else:
        print(f"Chỉ số cột {k} không hợp lệ. Mảng có {mang_2d.shape[1]} cột.")

print("--- c) Xuất các phần tử thuộc cột k: ---")
k_cot = 2 # Cột thứ 3 (chỉ số 2)
xuat_cot_k(array_a, k_cot)
print()

# d) Tìm dòng có tổng lớn nhất.
def tim_dong_tong_lon_nhat(mang_2d):
    """
    Tìm dòng có tổng các phần tử lớn nhất trong mảng 2 chiều.
    Trả về chỉ số dòng và tổng lớn nhất.
    """
    if mang_2d.size == 0:
        return -1, 0

    tong_cac_dong = np.sum(mang_2d, axis=1)
    chi_so_dong_max = np.argmax(tong_cac_dong)
    tong_max = tong_cac_dong[chi_so_dong_max]

    print(f"Dòng có tổng lớn nhất là dòng {chi_so_dong_max} với tổng là: {tong_max}")
    print(f"Các phần tử của dòng đó: {mang_2d[chi_so_dong_max]}")
    return chi_so_dong_max, tong_max

print("--- d) Tìm dòng có tổng lớn nhất: ---")
tim_dong_tong_lon_nhat(array_a)
print()

# e) Tìm cột có tích nhỏ nhất.
def tim_cot_tich_nho_nhat(mang_2d):
    """
    Tìm cột có tích các phần tử nhỏ nhất trong mảng 2 chiều.
    Trả về chỉ số cột và tích nhỏ nhất.
    """
    if mang_2d.size == 0:
        return -1, 1 # Tích của mảng rỗng có thể coi là 1

    tich_cac_cot = np.prod(mang_2d, axis=0)
    chi_so_cot_min = np.argmin(tich_cac_cot)
    tich_min = tich_cac_cot[chi_so_cot_min]

    print(f"Cột có tích nhỏ nhất là cột {chi_so_cot_min} với tích là: {tich_min}")
    print(f"Các phần tử của cột đó: {mang_2d[:, chi_so_cot_min]}")
    return chi_so_cot_min, tich_min

print("--- e) Tìm cột có tích nhỏ nhất: ---")
tim_cot_tich_nho_nhat(array_a)
print()

# f) Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a.
def xuat_dong_chan_cot_le(mang_2d):
    """
    Xuất các phần tử thuộc dòng chẵn và cột lẻ trong mảng 2 chiều.
    """
    print("Các phần tử thuộc dòng chẵn và cột lẻ:")
    found_elements = False
    for i in range(mang_2d.shape[0]):
        if i % 2 == 0:  # Dòng chẵn (0, 2, 4,...)
            for j in range(mang_2d.shape[1]):
                if j % 2 != 0:  # Cột lẻ (1, 3, 5,...)
                    print(f"  Phần tử tại [{i}, {j}]: {mang_2d[i, j]}")
                    found_elements = True
    if not found_elements:
        print("  Không tìm thấy phần tử nào thỏa mãn điều kiện.")

print("--- f) Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a: ---")
xuat_dong_chan_cot_le(array_a)
print()

# g) Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a.
def trung_binh_cong_chan_dong_le(mang_2d):
    """
    Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của mảng 2 chiều.
    """
    tong_chan = 0
    dem_chan = 0

    for i in range(mang_2d.shape[0]):
        if i % 2 != 0:  # Dòng lẻ (1, 3, 5,...)
            for j in range(mang_2d.shape[1]):
                if mang_2d[i, j] % 2 == 0:  # Phần tử chẵn
                    tong_chan += mang_2d[i, j]
                    dem_chan += 1

    if dem_chan > 0:
        trung_binh = tong_chan / dem_chan
        print(f"Trung bình cộng các phần tử chẵn thuộc dòng lẻ là: {trung_binh}")
        return trung_binh
    else:
        print("Không có phần tử chẵn nào thuộc dòng lẻ.")
        return 0

print("--- g) Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a: ---")
trung_binh_cong_chan_dong_le(array_a)
print()

# h) Tính trung bình cộng các phần tử thuộc biên.
def trung_binh_cong_phan_tu_bien(mang_2d):
    """
    Tính trung bình cộng các phần tử thuộc biên của mảng 2 chiều.
    """
    if mang_2d.size == 0:
        print("Không có phần tử biên (mảng rỗng).")
        return 0

    m, n = mang_2d.shape
    tong_bien = 0
    dem_bien = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m - 1 or j == 0 or j == n - 1: # Phần tử thuộc biên
                tong_bien += mang_2d[i, j]
                dem_bien += 1

    if dem_bien > 0:
        trung_binh = tong_bien / dem_bien
        print(f"Trung bình cộng các phần tử thuộc biên là: {trung_binh}")
        return trung_binh
    else:
        print("Không có phần tử biên (mảng quá nhỏ).")
        return 0

print("--- h) Tính trung bình cộng các phần tử thuộc biên: ---")
trung_binh_cong_phan_tu_bien(array_a)
print()
# i) tính trung bình các phần tử kh thuộc biên
def trung_binh_tich_phan_tu_khong_bien(mang_2d):
    """
    Tính trung bình tích các phần tử không thuộc biên của mảng 2 chiều.
    """
    if mang_2d.size == 0:
        print("Không có phần tử (mảng rỗng).")
        return 0

    m, n = mang_2d.shape
    phan_tu_khong_bien = []

    for i in range(m):
        for j in range(n):
            # Kiểm tra nếu phần tử không nằm trên biên
            if i > 0 and i < m - 1 and j > 0 and j < n - 1:
                phan_tu_khong_bien.append(mang_2d[i, j])

    if not phan_tu_khong_bien:
        print("Không có phần tử nào không thuộc biên (mảng quá nhỏ).")
        return 0
    else:
        # Tính tích của các phần tử không thuộc biên
        tich = np.prod(phan_tu_khong_bien)
        # Tính trung bình (hiểu là tích chia cho số lượng phần tử đó)
        # Tuy nhiên, đề bài có vẻ hơi không rõ ràng ở 'trung bình tích'.
        # Nếu hiểu là tích các phần tử không biên, ta sẽ in ra tích.
        # Nếu muốn trung bình, thì phải là trung bình cộng.
        # Giả sử ý của bạn là 'Tích của các phần tử không thuộc biên'.
        print(f"Các phần tử không thuộc biên: {phan_tu_khong_bien}")
        print(f"Tích các phần tử không thuộc biên là: {tich}")
        # Nếu thực sự muốn trung bình cộng của các phần tử không biên:
        # trung_binh_cong = np.mean(phan_tu_khong_bien)
        # print(f"Trung bình cộng các phần tử không thuộc biên là: {trung_binh_cong}")
        return tich # Trả về tích

print("--- i) Tính trung bình tích các phần tử không thuộc biên: ---")
trung_binh_tich_phan_tu_khong_bien(array_a)
print()
```

```python
#Bài 15.Một sinh viên gồm các thông tin sau:
#Mã sinh viên là chuỗi có 10 ký tự
#Tên là chuỗi tối đa 20 ký tự
#Năm sinh là một số nguyên
#Điểm trung bình là một số thực a.
import datetime

# a) Xây dựng cấu trúc SINHVIEN mô tả một sinh viên
class SINHVIEN:
    def __init__(self, ma_sv, ten_sv, nam_sinh, diem_trung_binh):
        # Mã sinh viên là chuỗi có 10 ký tự
        if not isinstance(ma_sv, str) or len(ma_sv) != 10:
            raise ValueError("Mã sinh viên phải là chuỗi có đúng 10 ký tự.")
        self.ma_sv = ma_sv

        # Tên là chuỗi tối đa 20 ký tự
        if not isinstance(ten_sv, str) or len(ten_sv) > 20:
            raise ValueError("Tên sinh viên phải là chuỗi và tối đa 20 ký tự.")
        self.ten_sv = ten_sv

        # Năm sinh là một số nguyên
        if not isinstance(nam_sinh, int) or nam_sinh < 1900 or nam_sinh > 2024: # Giới hạn năm sinh hợp lý
            raise ValueError("Năm sinh phải là một số nguyên hợp lệ.")
        self.nam_sinh = nam_sinh

        # Điểm trung bình là một số thực
        if not isinstance(diem_trung_binh, (int, float)) or not (0 <= diem_trung_binh <= 10):
            raise ValueError("Điểm trung bình phải là một số thực từ 0 đến 10.")
        self.diem_trung_binh = float(diem_trung_binh)

    def __str__(self):
        return (f"Mã SV: {self.ma_sv}, Tên: {self.ten_sv}, Năm sinh: {self.nam_sinh}, "
                f"ĐTB: {self.diem_trung_binh:.2f}")

# Ví dụ về cách tạo một đối tượng SINHVIEN
try:
    sv1 = SINHVIEN("01DH000001", "Nguyễn Văn A", 2003, 7.5)
    print(f"Sinh viên 1: {sv1}")
except ValueError as e:
    print(f"Lỗi tạo sinh viên: {e}")

print("\n" + "="*30 + "\n")

# b) Cho một mảng có n sinh viên.
danh_sach_sinh_vien = [
    SINHVIEN("01DH000001", "Nguyễn Võ Trọng Nhân", 2006, 7.5),
    SINHVIEN("02CD000002", "Trần Thị Lan", 2002, 4.8),
    SINHVIEN("03DH000003", "Phạm Minh Huy", 2001, 8.1),
    SINHVIEN("04CD000004", "Lê Thanh Mai", 2004, 6.0),
    SINHVIEN("05DH000005", "Hoàng Văn Lan", 2003, 5.0),
    SINHVIEN("06TH000006", "Phan Hữu Dũng", 2000, 9.2),
    SINHVIEN("07DH000007", "Nguyễn Thị Lan", 2003, 4.5)
]

print("Danh sách sinh viên hiện có:")
for sv in danh_sach_sinh_vien:
    print(f"- {sv}")

print("\n" + "="*30 + "\n")

# c) Viết hàm cho biết có bao sinh viên đủ điều kiện lên lớp.
def dem_sv_du_dieu_kien_len_lop(danh_sach_sv):
    """
    Đếm số sinh viên đủ điều kiện lên lớp (ĐTB >= 5).
    """
    so_luong = 0
    for sv in danh_sach_sv:
        if sv.diem_trung_binh >= 5.0:
            so_luong += 1
    return so_luong

print("--- c) Đếm sinh viên đủ điều kiện lên lớp: ---")
so_sv_du_dieu_kien = dem_sv_du_dieu_kien_len_lop(danh_sach_sinh_vien)
print(f"Số sinh viên đủ điều kiện lên lớp: {so_sv_du_dieu_kien}")

print("\n" + "="*30 + "\n")

# d) Xuất các sinh viên đủ 20 tuổi.
def xuat_sv_du_20_tuoi(danh_sach_sv):
    """
    Xuất danh sách sinh viên đủ 20 tuổi (tính theo năm hiện tại).
    """
    nam_hien_tai = datetime.datetime.now().year
    sv_du_20_tuoi = []
    print("Các sinh viên đủ 20 tuổi:")
    found = False
    for sv in danh_sach_sv:
        tuoi = nam_hien_tai - sv.nam_sinh
        if tuoi >= 20:
            print(f"- {sv} (Tuổi: {tuoi})")
            sv_du_20_tuoi.append(sv)
            found = True
    if not found:
        print("Không có sinh viên nào đủ 20 tuổi trong danh sách.")
    return sv_du_20_tuoi

print("--- d) Xuất các sinh viên đủ 20 tuổi: ---")
xuat_sv_du_20_tuoi(danh_sach_sinh_vien)

print("\n" + "="*30 + "\n")

# e) Đếm số sinh viên học hệ đại học.
def dem_sv_he_dai_hoc(danh_sach_sv):
    """
    Đếm số sinh viên học hệ đại học (mã sinh viên chứa 'DH' ở vị trí 2,3).
    """
    so_luong = 0
    for sv in danh_sach_sv:
        # Kiểm tra vị trí 2,3 (chỉ số 2 và 3 trong chuỗi 0-indexed)
        if len(sv.ma_sv) >= 4 and sv.ma_sv[2:4] == "DH": # Sửa từ [1:3] sang [2:4]
            so_luong += 1
    return so_luong

print("--- e) Đếm số sinh viên học hệ đại học: ---")
so_sv_dai_hoc = dem_sv_he_dai_hoc(danh_sach_sinh_vien)
print(f"Số sinh viên hệ đại học: {so_sv_dai_hoc}")

print("\n" + "="*30 + "\n")

# f) Cho biết trong mảng có bao nhiêu sinh viên có tên «Lan»
def dem_sv_ten_lan(danh_sach_sv):
    """
    Đếm số sinh viên có tên 'Lan' (tên cuối cùng).
    """
    so_luong = 0
    for sv in danh_sach_sv:
        # Tách tên ra và lấy phần tử cuối cùng (tên chính)
        # .split() sẽ tự động xử lý nhiều khoảng trắng và bỏ qua khoảng trắng ở đầu/cuối
        ten_cuoi = sv.ten_sv.strip().split(' ')[-1]
        if ten_cuoi == "Lan":
            so_luong += 1
    return so_luong

print("--- f) Đếm số sinh viên có tên 'Lan': ---")
so_sv_ten_lan = dem_sv_ten_lan(danh_sach_sinh_vien)
print(f"Số sinh viên có tên 'Lan': {so_sv_ten_lan}")

print("\n" + "="*30 + "\n")

# g) Cho biết trong mảng có bao nhiêu sinh viên có họ «Phan»
def dem_sv_ho_phan(danh_sach_sv):
    """
    Đếm số sinh viên có họ 'Phan'.
    """
    so_luong = 0
    for sv in danh_sach_sv:
        # Tách tên và lấy phần tử đầu tiên (họ)
        ho_sv = sv.ten_sv.strip().split(' ')[0]
        if ho_sv == "Phan":
            so_luong += 1
    return so_luong

print("--- g) Đếm số sinh viên có họ 'Phan': ---")
so_sv_ho_phan = dem_sv_ho_phan(danh_sach_sinh_vien)
print(f"Số sinh viên có họ 'Phan': {so_sv_ho_phan}")
print("\n" + "="*30 + "\n")
```

```python
#BTVN
#Bài 1
def tron_mang_min(a, b):
    # Lấy min của các phần tử tương ứng
    res = [min(x, y) for x, y in zip(a, b)]

    # Kiểm tra xem mảng nào dài hơn thì cộng phần dư vào
    if len(a) > len(b):
        res.extend(a[len(b):])
    elif len(b) > len(a):
        res.extend(b[len(a):])

    return res

# Test
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print("Mảng kết quả:", tron_mang_min(a, b))
```

```python
#Bài 2
# a. Tổng tam giác trên đường chéo phụ (kể cả đường chéo phụ)
def tong_tam_giac_tren_phu(a, n):
    total = 0
    for i in range(n):
        for j in range(n - i):
            total += a[i][j]
    return total

# b. Chuyển âm thành trị tuyệt đối
def tri_tuyet_doi(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            a[i][j] = abs(a[i][j])

# c. Thay chẵn bằng x
def thay_chan_bang_x(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x

# d. Kiểm tra toàn chẵn
def kiem_tra_toan_chan(a):
    for row in a:
        for val in row:
            if val % 2 != 0:
                return False
    return True

# e. Kiểm tra đối xứng (a[i][j] == a[j][i])
def la_doi_xung(a, n):
    for i in range(n):
        for j in range(i):
            if a[i][j] != a[j][i]:
                return False
    return True

# f. Đường chéo chính tăng dần
def duong_cheo_chinh_tang(a, n):
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            return False
    return True

# g. Xuất tam giác dưới đường chéo phụ
def xuat_tam_giac_duoi_phu(a, n):
    for i in range(n):
        for j in range(n - 1 - i, n):
            print(a[i][j], end=" ")
        print()

# h. Đường chéo phụ giảm dần
def duong_cheo_phu_giam(a, n):
    for i in range(n - 1):
        # i + j = n - 1 => j = n - 1 - i
        if a[i][n-1-i] <= a[i+1][n-2-i]:
            return False
    return True
```
